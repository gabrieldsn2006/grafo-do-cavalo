```python
from graph import Graph
from cc import ConnectedComponents
from breadth_first_paths import BreadthFirstPaths
from cycle import CycleDetector
```

## Carregando grafo e exibindo lista de adjacência

```python
def load_graph_from_file(file_path):
    with open(file_path, 'rt') as f:
        v = int(f.readline())
        e = int(f.readline())
        G = Graph(v)
        for line in f.readlines():
            if not line.startswith('#'):
                parts = line.split()
                if len(parts) == 2:
                    G.add_edge(int(parts[0]), int(parts[1]))
    return G

G = load_graph_from_file('../dados/entrada.txt')

print(f'Lista de Adjacência:\n{G}')
```

## Componentes do grafo

```python
cc = ConnectedComponents(G)
components = cc.get_all_components()

print(f'Nº de componentes conexas: {cc.count}')
for i in range(cc.count):
    print(f'Componente {i}: {components[i]}')
```

## Distância mínima entre duas posições

```python
# Parâmetros
source = 0
target = 2

bfs = BreadthFirstPaths(G, source)

if bfs.has_path_to(target):
    path = bfs.path_to(target)
    distance = len(path) - 1
    print(f"Menor caminho de {source} para {target}: {path}")
    print(f"Distância mínima: {distance}")
else:
    print(f"Não existe caminho entre {source} e {target} no grafo")
```

## Cliclos do grafo

```python
cyc = CycleDetector(G)
print("Número de ciclos:", cyc.num_cycles())
print("Ciclos:", cyc.cycles())
```
